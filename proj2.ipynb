```python
plt.figure(figure=(8,6))
explode=[0,0.1]
colors=['blue','pink']
plt.pie(df['Survived'].value_counts(), labels=df['Survived'].value_counts().index, autopct='%1.1f%%',explode=explode,colors=colors)
plt.title('Survival rate')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df,x='Pclass',y='Age',palette='Blues')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='coolwarm')
plt.title('Pair Plot of Numerical Variables')
plt.show()
```
